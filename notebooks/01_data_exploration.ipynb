```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

### Chargement du dataset

```python
# Style des graphiques
sns.set(style="whitegrid", palette="muted", font_scale=1.1)

df = pd.read_csv("../data/raw/output.csv")  # Remplace par ton fichier
df.head()
```

```python
# Aperçu global
print("Dimensions du dataset :", df.shape)
df.info()
df.describe(include="all").T
```

### Distribution des prix

```python
plt.figure(figsize=(8,5))
sns.histplot(df['price'], kde=True, bins=40, color="royalblue")
plt.title("Distribution des prix des maisons")
plt.show()

# Statistiques de base sur les prix
df['price'].describe()
```

```python
#Matrice de correlation entre les variables numériques

plt.figure(figsize=(25, 20))
sns.heatmap(df.corr(numeric_only=True), annot = True)
plt.show()
```

### Identifier les relations avec le prix

```python
# Corrélation numérique
corr = df.corr(numeric_only=True)
plt.figure(figsize=(12,8))
sns.heatmap(corr[['price']].sort_values(by='price', ascending=False), 
            annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Corrélation des variables avec le prix")
plt.show()

# Détection des maisons très grandes par rapport au prix
plt.figure(figsize=(8,5))
sns.scatterplot(x='sqft_living', y='price', data=df)
plt.ylim(0, 3_000_000)
plt.xlim(0, 10_000)
plt.title("Prix vs Surface habitable")
plt.show()

```

### Valeurs aberrantes

```python

for col in df.select_dtypes('number').columns:
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))
    sns.histplot(df[col], ax = axes[0])
    sns.boxplot(df[col], ax = axes[1], showmeans=True)
plt.show() 
```

```python
# Prix moyen par ville
top_cities = df.groupby("city")['price'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_cities.values, y=top_cities.index, palette="viridis")
plt.title("Prix moyen par ville (Top 10)")
plt.xlabel("Prix moyen")
plt.show()
```
